```python
#Importando as bibliotecas
import random
import matplotlib.pyplot as plt

#Gerar uma lista com 10 númenros aleatórios entre 1 e 100
dados = [random.randint(1, 100) for _ in range(10)]

#Exibir a lista gerada
print("Lista de dados: ", dados)

#Plotar os dados
plt.plot(dados, marker='o') #Marcador em cada ponto para melhor visualização
plt.title("Gráfico dos dados")
plt.xlabel("Índice")
plt.ylabel("Valor")
plt.grid(True) #Grid para melhor visualização
plt.show()
```
